# ノイラミニダーゼ

```python
%%writefile fig.pml
reinitialize
fetch 1bji
color red, ss h
color yellow, ss s
color green, ss l+''
symexp sym, 1bji, (1bji), 6
disable sym08000000 or sym16000000 or sym44000000
set_view (\
     0.977228284,   -0.210333690,   -0.027956437,\
     0.211222947,    0.976846874,    0.033979051,\
     0.020163115,   -0.039110661,    0.999030232,\
     0.000005066,   -0.000012636, -320.043151855,\
    -1.688617706,   -1.803961277,   49.017486572,\
   221.435028076,  418.651245117,  -20.000000000 )
png fig.png
```
